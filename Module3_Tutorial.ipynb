Given an array arr[] containing only 0s, 1s, and 2s. Sort the array in ascending order.
Note: You need to solve this problem without utilizing the built-in sort function.

```python
class Solution:
    def sort012(self, arr, n):
        low, mid, high = 0, 0, n - 1

        while mid <= high:
            if arr[mid] == 0:
                arr[low], arr[mid] = arr[mid], arr[low]
                low += 1
                mid += 1
            elif arr[mid] == 1:
                mid += 1
            else:  # arr[mid] == 2
                arr[mid], arr[high] = arr[high], arr[mid]
                high -= 1

if __name__ == "__main__":
    arr = [2, 1, 2, 2, 2, 1, 0]
    n = len(arr)
    ob = Solution()
    ob.sort012(arr, n)
    print(arr)  # Output: [0, 1, 1, 2, 2, 2, 2]
```

```python
Given a sorted array arr[] and an integer k, find the position(0-based indexing) at which k is present in the array using binary search. If k doesn't exist in arr[] return -1. 

Note: If multiple occurrences are there, please return the smallest index.
```

```python
class Solution:
    def binarysearch(self, arr, k):
        low = 0
        high = len(arr) - 1
        ans = -1
        
        while low <= high:
            mid = (low + high) // 2
            
            if arr[mid] == k:
                ans = mid            # store position
                high = mid - 1       # keep searching LEFT for first occurrence
                
            elif arr[mid] < k:
                low = mid + 1
            
            else:
                high = mid - 1
        
        return ans

```
